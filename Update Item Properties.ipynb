```python
import arcgis
from arcgis import GIS
```

### Connect to AGOL

```python
gis = GIS('Pro')
```

### Search for prop studies

```python
query = '(title: _Lv12 OR title: _60) AND owner:mechomicky@burnsmcd.com_bmcd_gis'
search = gis.content.advanced_search(query)
search
```

```python
content = search['results']
content
```

```python
while search['start'] + search['num'] < search['total']:
    nextStart = search['start'] + search['num']
    print('retrieving items starting at ' + str(nextStart))
    search = gis.content.advanced_search(query, start=nextStart)
    content += search['results']
    
```

### Add RadioCompass tag to content

```python
for item in content:
    item.update(item_properties={'tags':'RadioCompass'})
```
